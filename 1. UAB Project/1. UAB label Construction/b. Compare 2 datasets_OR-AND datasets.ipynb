```python
def compare_2_columns(df1,df2,l1,l2):
    result1 = []
    result2 = []
    first_list = df1.iloc[:,l1]    ### pandas series
    second_list =df2.iloc[:,l2]    ### pandas series
    
    for i in range(len(df1.iloc[:,l1])):
        if  first_list[i] ==1 and second_list[i] ==1:
            result1.append(1)
            result2.append(1)
        elif first_list[i] ==0 and second_list[i] ==0:
            result1.append(0)
            result2.append(0)
        #elif first_list[i] ==np.NaN and second_list[i] ==np.NaN: ##
        #    result1.append(2)  ##
        #    result2.append(2)  ##
        else:
            result1.append(1)
            result2.append(0)
    return (result1,result2) 
```

```python
def compare_2_df(df1,df2):
    p = len(df1.columns)
    df3 = []
    df4 = []
    for j in range(p):
        Result1 = compare_2_columns(df1,df2,j,j)
        one = Result1[0]
        two = Result1[1]
        df3.append(one)
        df4.append(two)
    
    return (df3,df4)
```

```python
df_OR = pd.DataFrame(compare_2_df(df11, df22)[0])
df_OR = pd.DataFrame.transpose(df_OR)
```

```python
df_AND = pd.DataFrame(compare_2_df(df11, df22)[1])
df_AND = pd.DataFrame.transpose(df_AND)
```
