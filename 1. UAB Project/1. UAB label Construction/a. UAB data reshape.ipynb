# Step 1.1: select columns
- make 2 lists for **grader 1** & **grader 2**

```python
grader1 = col_names[0:45]
print(len(grader1))
print(grader1)
```

```python
from itertools import chain
grader2 = list(chain([col_names[0]], col_names[45:89]))
print(len(grader2))
print(grader2)
```

# Step 1.2: select columns
- each grader is being slitted into 2 lists
- 4 lists in total

```python
a1 = grader1[0:4]
b1 = grader1[4:24]
c1 = grader1[24:44]
grader1_od = a1 + b1
grader1_os = a1 + c1
```

```python
a2 = grader2[0:4]
b2 = grader2[4:24]
c2 = grader2[24:44]
grader2_od = a2 + b2
grader2_os = a2 + c2
```

# ==========================================================

```python
# list 1: grader1 - od
df_grader1_od = df[grader1_od]
copy = df_grader1_od["eye"]
df_grader1_od.insert(3, "OD_flag", copy)
df_grader1_od.insert(4, "Eye", copy)

df_grader1_od['OD_flag'].replace(3,'od', inplace = True)  ####
df_grader1_od['OD_flag'].replace(1,'od', inplace = True)  ####
df_grader1_od['OD_flag'].replace(2,'od', inplace = True)  ########

df_grader1_od['Eye'].replace(3,'Right', inplace = True)  ####
df_grader1_od['Eye'].replace(1,'Right', inplace = True)  ####
df_grader1_od['Eye'].replace(2,'Right', inplace = True)  ########

df_grader1_od.drop('eye', inplace=True, axis=1)

print(df_grader1_od.shape)
df_grader1_od.head()
```

```python
# list 2: grader1 - os
df_grader1_os = df[grader1_os]
copy = df_grader1_os["eye"]
df_grader1_os.insert(3, "OD_flag", copy)
df_grader1_os.insert(4, "Eye", copy)

df_grader1_os['OD_flag'].replace(3,'os', inplace = True)  ####
df_grader1_os['OD_flag'].replace(2,'os', inplace = True)  ####
df_grader1_os['OD_flag'].replace(1,'os', inplace = True)  ########

df_grader1_os['Eye'].replace(3,'Left', inplace = True)  ####
df_grader1_os['Eye'].replace(2,'Left', inplace = True)  ####
df_grader1_os['Eye'].replace(1,'Left', inplace = True)  ########

df_grader1_os.drop('eye', inplace=True, axis=1)

print(df_grader1_os.shape)
df_grader1_os.head()
```

```python
# list 3: grader2 - od
df_grader2_od = df[grader2_od]
copy = df_grader2_od["eye_v2"]
df_grader2_od.insert(3, "OD_flag", copy)
df_grader2_od.insert(4, "Eye", copy)

df_grader2_od['OD_flag'].replace(3,'od', inplace = True)  ####
df_grader2_od['OD_flag'].replace(1,'od', inplace = True)  ####
df_grader2_od['OD_flag'].replace(2,'od', inplace = True)  ########

df_grader2_od['Eye'].replace(3,'Right', inplace = True)  ####
df_grader2_od['Eye'].replace(1,'Right', inplace = True)  ####
df_grader2_od['Eye'].replace(2,'Right', inplace = True)  ########

df_grader2_od.drop('eye_v2', inplace=True, axis=1)

print(df_grader2_od.shape)
df_grader2_od.head()
```

```python
# list 4: grader2 - os
df_grader2_os = df[grader2_os]
copy = df_grader2_os["eye_v2"]
df_grader2_os.insert(3, "OD_flag", copy)
df_grader2_os.insert(4, "Eye", copy)

df_grader2_os['OD_flag'].replace(3,'os', inplace = True)  ####
df_grader2_os['OD_flag'].replace(2,'os', inplace = True)  ####
df_grader2_os['OD_flag'].replace(1,'os', inplace = True)  ########

df_grader2_os['Eye'].replace(3,'Left', inplace = True)  ####
df_grader2_os['Eye'].replace(2,'Left', inplace = True)  ####
df_grader2_os['Eye'].replace(1,'Left', inplace = True)  ########

df_grader2_os.drop('eye_v2', inplace=True, axis=1)

print(df_grader2_os.shape)
df_grader2_os.head()
```

```python
df_grader1_od.to_csv('/Users/yuhuicao/Desktop/UAB label/1. UAB label reshape grader1-od.csv', index = False)
df_grader1_os.to_csv('/Users/yuhuicao/Desktop/UAB label/2. UAB label reshape grader1-os.csv', index = False)
df_grader2_od.to_csv('/Users/yuhuicao/Desktop/UAB label/3. UAB label reshape grader2-od.csv', index = False)
df_grader2_os.to_csv('/Users/yuhuicao/Desktop/UAB label/4. UAB label reshape grader2-os.csv', index = False)
```
