# Pairwise Plot

# 1. Import Data

```python
import pandas as pd
```

```python
Grader1_labels_VF = pd.read_csv("11. Grader1_labels_VF.csv")
print("The shape of Grader1_labels_VF is ", Grader1_labels_VF.shape)
Grader1_labels_VF.head()
```

```python
Grader1_labels_VF.columns[:35]
```

```python
Grader1_labels_VF.columns[29:]
```

```python
l1 = Grader1_labels_VF.columns[29:]
```

```python
vf1 = Grader1_labels_VF[l1]
print("The shape of vf is", vf1.shape)
vf1.head()
```

```python
l2 = Grader1_labels_VF.columns[5:25]

vf2 = Grader1_labels_VF[l2]
print("The shape of vf is", vf2.shape)
vf2.head()
```

```python
vf = pd.concat([vf1, vf2], axis = 1)

vf['ptid'] = Grader1_labels_VF['ptid']
vf['Eye'] = Grader1_labels_VF['Eye']

print("The shape of vf is", vf.shape)
vf.head()
```

```python
print("Contains NaN?", vf.isnull().values.any())
print("Number of NaN:", vf.isnull().sum().sum())
```

# 2. "3-feature" dataset (MD, PSD, VFI) 
# vs. pairwise plot

```python
vf_3features = vf[vf.columns[:3]]
print(vf_3features.shape)
vf_3features['sad'] = vf['sad']
vf_3features.head()
```

```python
import seaborn as sns 
sns.pairplot(vf_3features, hue="sad")
```

### Summary 1: MD, PSD, VFI cannot distinguish "sad" very well.  T_T

```python
vf_3features = vf[vf.columns[:3]]
vf_3features['iad'] = vf['iad']
sns.pairplot(vf_3features, hue="iad")
```

### Summary 2: MD, PSD, VFI cannot distinguish "iad" very well. T_T

```python
vf_3features = vf[vf.columns[:3]]
vf_3features['sns'] = vf['sns']
sns.pairplot(vf_3features, hue="sns")
```

### Summary 3: MD, PSD, VFI cannot distinguish "sns" very well. T_T

```python
vf_3features = vf[vf.columns[:3]]
vf_3features['normal'] = vf['normal']
sns.pairplot(vf_3features, hue="normal")
```

### Summary 4: MD, PSD, VFI cannot distinguish "normal" very well. T_T

```python
vf_3features = vf[vf.columns[:3]]
vf_3features['dad'] = vf['dad']
sns.pairplot(vf_3features, hue="dad")
```

### Summary 5: MD, PSD, VFI can distinguish "dad" very well. ^_^

```python
vf_3features = vf[vf.columns[:3]]
vf_3features['ci'] = vf['ci']
sns.pairplot(vf_3features, hue="ci")
```

### Summary 6: MD, PSD, VFI can distinguish "ci" very well. ^_^

### Concolusion: 
- MD, PSD and VFI, 3 features might be able to distiguish "ci" and "dad" very well.
