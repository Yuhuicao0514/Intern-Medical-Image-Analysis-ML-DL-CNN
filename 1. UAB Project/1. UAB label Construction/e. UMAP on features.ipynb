# 1. UMAP reduces down to 2D.
# 265 Features:
- MD (1)
- PSD (1)

- TD (52)

```python
1 + 1 + 1 + 54 + 52*4
```

```python
vf_umap = vf[vf.columns[:265]]
print(vf_umap.shape)
vf_umap.head()
```

```python
import umap
```

```python
reducer = umap.UMAP()
```

```python
embedding = reducer.fit_transform(vf_umap)
embedding.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=[sns.color_palette()[x] for x in vf.normal.map({0.0:0, 1.0:1})])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection: UAB - 265 features - label "normal"')
```

```python
plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=[sns.color_palette()[x] for x in vf.ci.map({0.0:0, 1.0:1})])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection: UAB - 265 features - label "ci"')
```

```python
plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=[sns.color_palette()[x] for x in vf.dad.map({0.0:0, 1.0:1})])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection: UAB - 265 features - label "dad"')
```

```python
plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=[sns.color_palette()[x] for x in vf.sad.map({0.0:0, 1.0:1})])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection: UAB - 265 features - label "sad"')
```

```python
plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=[sns.color_palette()[x] for x in vf.iad.map({0.0:0, 1.0:1})])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection: UAB - 265 features - label "iad"')
```
