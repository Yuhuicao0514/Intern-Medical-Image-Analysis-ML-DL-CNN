# Combining (merge) Visual Field Data

```python
import pandas as pd
```

```python
label_g1 = pd.read_csv("7. UAB label reshape grader1 (drop NaN).csv")
label_g2 = pd.read_csv("8. UAB label reshape grader2 (drop NaN).csv")

label_OR = pd.read_csv("9. UAB label reshape OR (drop NaN).csv")
label_AND = pd.read_csv("10. UAB label reshape AND (drop NaN).csv")

vff_old = pd.read_csv("20210809_vf_guided_UAB_raw_perimetry_data.csv")
vff = pd.read_csv("20220708_vf_guided_UAB_raw_perimetry_data.csv")
```

```python
print("The label of label_g1: ", label_g1.shape)
print("The label of label_g2: ", label_g2.shape)
print("The label of label_OR: ", label_OR.shape)
print("The label of label_AND: ", label_AND.shape)
print("The label of vf (old) data: ", vff_old.shape)
print("The label of vf (new) data: ", vff.shape)
```

```python
vf1 = vff[vff.columns[143:413]]
print("The shape of vf1 is: ", vf1.shape)
vf1.head()
```

```python
vff.head()
```

```python
l = ["ptid", "Eye", "BirthDate", "Age", "TestStrategy", "TestPattern", "MD", "PSD", "VFI"]
vf2 = vff[l]
print("The shape of vf2 is: ", vf2.shape)
vf2.head()
```

```python
vf = pd.concat([vf2, vf1], axis = 1)
print("The shape of vf is: ", vf.shape)
vf.head()
```

```python
print("The shape of label_g1: ", label_g1.shape)
print("The shape of label_g2: ", label_g2.shape)

print("The shape of label_OR: ", label_OR.shape)
print("The shape of label_AND: ", label_AND.shape)

print("The label of vf data: ", vf.shape)
```

```python
grader1 = pd.merge(label_g1, vf, on = ["ptid", 'Eye'])
print("The shape of grader1 is:", grader1.shape)
grader1.head()
```

```python
grader2 = pd.merge(label_g2, vf, on = ["ptid", 'Eye'])
print("The shape of grader2 is: ", grader2.shape)
grader2.head()
```

```python
OR = pd.merge(label_OR, vf, on = ["ptid", 'Eye'])
print("The shape of OR is: ", OR.shape)
OR.head()
```

```python
AND = pd.merge(label_AND, vf, on = ["ptid", 'Eye'])
print("The shape of AND is: ", AND.shape)
AND.head()
```

### Clean NaN: Drop 3 rows

```python
### drop rows (3)
grader1 = grader1.dropna(subset = ["Threshold_1"])
grader2 = grader2.dropna(subset = ["Threshold_1"])
OR = OR.dropna(subset = ["Threshold_1"])
AND = AND.dropna(subset = ["Threshold_1"])

print("The shape of grader1 is", grader1.shape)
print("The shape of grader2 is", grader2.shape)
print("The shape of OR is", OR.shape)
print("The shape of AND is", AND.shape)
```
