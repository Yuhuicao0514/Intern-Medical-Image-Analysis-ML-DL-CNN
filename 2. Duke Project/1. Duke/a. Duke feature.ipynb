# Part 1. Combine md_242 and md_302
### Create a new column, md

```python
import numpy as np
md = [0] * feature.shape[0]

for i in range(feature.shape[0]):
    if feature["md_242"].isnull()[i] == False:
        md[i] = feature["md_242"][i]
    else:
        md[i] = feature["md_302"][i]
len(md)
```

```python
feature.insert(32, "md", md)
print("The shape of feature is", feature.shape)
feature.head()
```

```python
print("Contains NaN?", feature["md_242"].isnull().values.any())
print("Number of NaN:", feature["md_242"].isnull().sum().sum())
```

```python
print("Contains NaN?", feature["md_302"].isnull().values.any())
print("Number of NaN:", feature["md_302"].isnull().sum().sum())
```

```python
print("Contains NaN?", feature["md"].isnull().values.any())
print("Number of NaN:", feature["md"].isnull().sum().sum())
```
